```python
from sqlalchemy import create_engine
from sqlalchemy import Table, Column, Integer, String, MetaData, ForeignKey
from sqlalchemy_utils import functions

if functions.database_exists('sqlite:///HoopStat.db'):
    functions.drop_database('sqlite:///HoopStat.db')

# Global Variables
SQLITE  = 'sqlite'

# Table Names
ACTIONTYPE = 'actionType'
PLAYERS = 'players'
TEAMS = 'teams'
GAMES = 'games'
EVENTS = 'events'

class MyDatabase:
    # http://docs.sqlalchemy.org/en/latest/core/engines.html
    DB_ENGINE = {
        SQLITE: 'sqlite:///HoopStat.db'
    }

    # Main DB Connection Ref Obj
    db_engine = None
    def __init__(self, dbtype, username='', password='', dbname=''):
        dbtype = dbtype.lower()
        if dbtype in self.DB_ENGINE.keys():
            engine_url = self.DB_ENGINE[dbtype].format(DB=dbname)
            self.db_engine = create_engine(engine_url)
            print(self.db_engine)
        else:
            print("DBType is not found in DB_ENGINE")
    def create_db_tables(self):
        metadata = MetaData()
        actionType = Table(ACTIONTYPE, metadata,
                      Column('id', Integer, primary_key=True)
                      )
        teams = Table(TEAMS,metadata,
                     Column('id',Integer, primary_key=True)
                     )
        games = Table(GAMES,metadata,
                      Column('id',String, primary_key=True),
                      Column('home', None, ForeignKey('teams.id')),
                      Column('homeScore',Integer),
                      Column('away', None, ForeignKey('teams.id')),
                      Column('awayScore',Integer),
                      Column('gameDate',String),
                      Column('isProcessed',Boolean,default = False)
                     )
        players = Table(PLAYERS, metadata,
                        Column('id', Integer, primary_key=True),
                        Column('name', String),
                        Column('team', None, ForeignKey('teams.id')),
                        Column('year', Integer)
                        )
        events = Table(EVENTS, metadata,
                       Column('id',Integer, primary_key=True),
                       Column('gameid',String),
                       Column('time',String),
                       Column('action',String),
                       Column('scoreHome',String),
                       Column('scoreAway',String),
                       Column('team',String),
                       Column('duration',String)
                      )
        
        try:
            metadata.create_all(self.db_engine)
            print("Tables created")
        except Exception as e:
            print("Error occurred during Table creation!")
            print(e)
    
    # Insert, Update, Delete
    def execute_query(self, query=''):
        if query == '' :
            return
        print (query)
        with self.db_engine.connect() as connection:
            try:
                connection.execute(query)
            except Exception as e:
                print(e)
    
dbms = MyDatabase(SQLITE, dbname='HoopStat.db')
dbms.create_db_tables()
```
